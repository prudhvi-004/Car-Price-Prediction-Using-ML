**Encoding Manually**

```python
#encoding fuel_type
car_data.replace({'fuel_type':{'Petrol':0,'Diesel':1,'CNG':2,'LPG':3,'Electric':4}},inplace=True)

#encoding seller type
car_data.replace({'seller_type':{'Dealer':0,'Individual':1,'Trustmark Dealer':2}},inplace=True)

#encoding Transmission type
car_data.replace({'transmission_type':{'Manual':0,'Automatic':1}},inplace=True)
```

**Splitting the Data**

```python
car_data = car_data.drop(columns=['Unnamed: 0','car_name','brand','model'])
```

```python
car_data.head()
```

```python
X = car_data.drop(columns=['selling_price'])
```

```python
print(X)
```

```python
Y = car_data['selling_price']
```

```python
print(Y)
```

**Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X_scaled.std()
```

```python
X_scaled.head()
```

```python
X = X_scaled
```

```python
X.head()
```

```python
X.shape
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Model Evaluation**

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
def evaluate(true,predicted):
  mae = mean_absolute_error(true,predicted)
  mse = mean_squared_error(true,predicted)
  r2 = r2_score(true,predicted)
  return mae,mse,r2
```

**Model Training**

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
models = [LinearRegression(),Lasso(),Ridge(),DecisionTreeRegressor(),RandomForestRegressor(),KNeighborsRegressor(),SVR(),AdaBoostRegressor(),GradientBoostingRegressor(),XGBRegressor()]
```

```python


for model in models:
  model.fit(X_train,Y_train)

  Y_train_pred = model.predict(X_train)
  Y_test_pred = model.predict(X_test)

  print("Model Name:",model)
  print()

  print("Training Results:")
  mae,mse,r2 = evaluate(Y_train,Y_train_pred)
  print("MAE:",mae)
  print("MSE:",mse)
  print("R2:",r2)

  print()

  print("Testing Result:")
  mae,mse,r2 = evaluate(Y_test,Y_test_pred)
  print("MAE:",mae)
  print("MSE:",mse)
  print("R2:",r2)

  print()
  print()

  print("-------------------------------------------------------------------------------------")
```

Selected Models KNN,RF,Gradient

**HyperParameter Tuning**

```python
# Hypermater Tuning
knn_params = {'n_neighbors': [2,3,10,20,40,50]}
rf_params = {
    "max_depth": [5, 8, 15, None, 10],
    "max_features": [5, 7, "sqrt", None],
    "min_samples_split": [2, 8, 15, 20],
    "n_estimators": [100, 200, 500, 1000]
}
# xg_params = {'learning_rate':[0.1,0.01],
#             'max_depth':[5,8,12,20,30],
#             'n_estimators':[100,200,300],
#             'colsample_bytree':[0.5,0.8,1,0.3,0.4]}

# gradient_params={
#     'loss': ['squared_error', 'absolute_error', 'huber', 'quantile'],
#     'criterion': ['friedman_mse', 'squared_error'],
#     'n_estimators': [100, 200, 500],
#     'max_depth': [5, 8, 15, None]
# }
```

# **Model Selection**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv_models = [('RF', RandomForestRegressor(), rf_params),
                   ('KNN', KNeighborsRegressor(), knn_params)
                  ]
```

```python
model_param = {}
for name, model, params in randomcv_models:
    print(f"\n🔍 Tuning {name}...")
    random = RandomizedSearchCV(
        estimator=model,
        param_distributions=params,
        n_iter=10,
        cv=2,
        verbose=2,
        n_jobs=-1,
        random_state=42
    )
    random.fit(X_train, Y_train)
    model_param[name] = random.best_params_, random.best_score_
    print(f"{name} has been successfully tuned.")

# Show best parameters
for model_name in model_param:
    print(f"\n-------------------- Best Params for {model_name} -------------------")
    print(model_param[model_name])
```

```python
model = RandomForestRegressor(max_depth=None, max_features='sqrt', min_samples_split=2,
                       n_estimators=100, random_state=42)
```

```python
print(model)
```

```python
model.fit(X_train,Y_train)

Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

print(r2_score(Y_train,Y_train_pred))
print(r2_score(Y_test,Y_test_pred))
```

```python
new_customer_features = [[11, 60000, 1, 0, 0, 17.00, 1197, 80.00, 5]]
new_customer_scaled = scaler.transform(new_customer_features)
prediction = model.predict(new_customer_scaled)
print(f"Predicted price: ₹{prediction[0]:,.2f}")
```
